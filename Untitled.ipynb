```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Read Data From CSV

```python
df = pd.read_pickle('dummy_data.pkl')
print(df)
```

Correlation Matrix of Raw Data:

```python
df.dropna(inplace=True,axis='columns')



corrMatrix = df.corr()
sns.heatmap(corrMatrix, annot=True)
fig = plt.gcf()
fig.set_size_inches(60, 60)
plt.show()
```
